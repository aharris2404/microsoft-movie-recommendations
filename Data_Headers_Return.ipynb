This file puts the column names of all the csv's into a nice dataframe so we can look at the data we have available in a single location.

```python
import os
import pandas as pd
```

```python
%cd data
!pwd
!ls #if this doesn't work just change into the directory FOO/microsoft-movie-recommendations/data
```

```python
DF_cols_list = list()
for file in os.listdir(): #grab names of files within the directory as a list and iterate over them
    if file[-6:-3] == 'csv': #in case file is a CSV
        DF_cols_list.append(([file] + list(pd.read_csv(file).columns))) #throw that boy in the df
    elif file[-6:-3] == 'tsv': #in case file is a tsv
        DF_cols_list.append(([file] + list(pd.read_csv(file, sep='\t', encoding='latin-1').columns))) #throw that boy in the df
    else:
        continue
```

```python
columns_in_each_csv = pd.DataFrame(DF_cols_list)
columns_in_each_csv
```
