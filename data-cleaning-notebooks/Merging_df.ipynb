```python
genre_df = pd.read_csv('../data/df1_cleaned_for_merge.csv') #basic
gross_df = pd.read_csv('../data/bom_imdb_revenue_concatenated.csv') #rev_data
ratings_df = pd.read_csv('../data/imdb.title.ratings.cleaned_for_merge.csv') #budget_data
```

```python
display(genre_df.head())
display(gross_df.head())
display(ratings_df.head())
```

```python
genre_df = genre_df.drop(['Unnamed: 0'], axis=1)
gross_df = gross_df.drop(['Unnamed: 0'], axis=1)
ratings_df = ratings_df.drop(['Unnamed: 0'], axis=1)
```

```python
display(genre_df.head())
display(gross_df.head())
display(ratings_df.head())
```

```python
full_df = ratings_df.merge(genre_df, on='tconst')
```

```python
full_df = full_df.drop(['Talk-Show', 'Adult', 'Game-Show', 'Short', 'News', 'Reality-TV', 'tconst', 'Musical', 'Western'], axis=1)
```

```python
full_gross = full_df.merge(gross_df, on='title')
```
